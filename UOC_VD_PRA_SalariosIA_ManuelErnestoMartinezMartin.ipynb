### Modificación de columnas

```python
# cambio de nombre en columnas
salarios_ia_df.rename(columns={
    'job_title': 'puesto_trabajo',
    'remote_ratio': 'porcentaje_remoto',
    'work_year': 'año_trabajo',
    'salary': 'salario',
    'salary_currency': 'moneda_salario',
    'salary_in_usd': 'salario_usd',
    'company_location': 'codigo_pais_empresa',
    'employee_residence': 'codigo_pais_residencia_empleado',
    'company_size': 'tamaño_empresa',
    'experience_level': 'nivel_experiencia',
    'employment_type': 'tipo_contrato'
}, inplace=True)
```

```python
# reemplazo de columnas
salarios_ia_df['nivel_experiencia'] = salarios_ia_df['nivel_experiencia'].replace(experience_level_map)
salarios_ia_df['tamaño_empresa'] = salarios_ia_df['tamaño_empresa'].replace(company_size_map)
salarios_ia_df['tipo_contrato'] = salarios_ia_df['tipo_contrato'].replace(employment_type_map)
```

### Nuevas columnas

```python
# Tasas de conversión de USD a EUR (media de cada año)
usd_to_eur_rates = {
    2020: 0.877,
    2021: 0.8458,
    2022: 0.951,
    2023: 0.925
}
```

```python
def convert_to_eur(dataframe):
    if (dataframe['moneda_salario'] == "EUR"):
        return dataframe['salario']
    tmp = dataframe['salario_usd'] * usd_to_eur_rates[dataframe['año_trabajo']]
    return int(round(tmp, -3))
```

```python
# nuevas columnas
salarios_ia_df['modalidad_trabajo'] = salarios_ia_df['porcentaje_remoto'].replace(remote_ratio_map)
salarios_ia_df['pais_empresa'] = salarios_ia_df['codigo_pais_empresa'].map(lambda x: country_map[x]['pais'])
salarios_ia_df['pais_residencia_empleado'] = salarios_ia_df['codigo_pais_residencia_empleado'].map(lambda x: country_map[x]['pais'])
salarios_ia_df['codigo_continente_empresa'] = salarios_ia_df['codigo_pais_empresa'].map(lambda x: country_map[x]['codigo_continente'])
salarios_ia_df['continente_empresa'] = salarios_ia_df['codigo_pais_empresa'].map(lambda x: country_map[x]['continente'])
salarios_ia_df['codigo_continente_residencia_empleado'] = salarios_ia_df['codigo_pais_residencia_empleado'].map(lambda x: country_map[x]['codigo_continente'])
salarios_ia_df['continente_residencia_empleado'] = salarios_ia_df['codigo_pais_residencia_empleado'].map(lambda x: country_map[x]['continente'])
salarios_ia_df['salario_eur'] = salarios_ia_df.apply(convert_to_eur, axis=1)
```

```python
bins = [0, 15, 20, 30, 40, 60, 100, 200, 300, 400, 500, 600, 1000]
labels = ['< 15 Mil', '15-20 Mil', '20-30 Mil', '30-40 Mil', '40-60 Mil', '60-100 Mil', '100-200 Mil', '200-300 Mil', '300-400 Mil', '400-500 Mil', '500-600 Mil', '> 600 Mil']
salarios_ia_df['rango_salarial_eur'] = pd.cut(salarios_ia_df['salario_eur'] / 1000, bins=bins, labels=labels, include_lowest=True, right=False)
```
